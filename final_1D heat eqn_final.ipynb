```python
# This code is to demonstrate the 1D heat equation
import numpy as np
import matplotlib.pyplot as plt
```

```python
length = 10 # length of the rod
k = 0.466 # heat constant of rod ( perticularly of steel)
# k = 0.01
t_left = 200 # temperature at left side of the rod
t_right = 200 #temperature at right side of rod
tot_time = 10  # time when simulation runs
dt = 0.001 # time step for dt
```

```python
dx = 0.1  #space between  points ( lesser the better)
x_array = np.linspace(0,length,int(length/dx)) # x array of points
# x_array
t_array = np.linspace(0,tot_time,int(tot_time/dt)) # t_array
# t_array
u = np.zeros([len(t_array), len(x_array)])
# u
```

```python
u[:, 0] = t_left     # u(t, 0) = 200
u[:, -1] = t_right   # u(t, length) = 200
```

```python
for t in range(1, len(t_array)-1):
    for x in range(1, len(x_array)-1):
        u[t+1, x] = k * (dt / dx**2) * (u[t, x+1] - 2*u[t, x] + 
                    u[t, x-1]) + u[t, x]
```

```python
# import matplotlib.pyplot as plt
plt.plot(x_array, u[-2], 'black')
# plt.pause(.001)
# plt.cla()
plt.plot(x_array, u[0])
plt.ylabel("Temperature (CËš)")
plt.xlabel("Distance Along Rod (m)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
xmin = min(x_array)  # Assuming x_array is defined in your wave equation solving code
xmax = max(x_array)
nx = len(x_array)


# Calculate nt based on the length of u or any other relevant consideration
nt = len(u)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))
plt.rcParams['animation.embed_limit'] = 50  


# Define a function to update the plot for each frame of the animation
def update(frame):
    ax.clear() # clears the axis before plotting the new frame.
    ax.plot(np.linspace(xmin, xmax, nx), u[frame, :], label=f"Time Step {frame}") # plots the wave at the current time step (frame).
    ax.set_xlabel('x')
    ax.set_ylabel('u')
    ax.set_title('Heat Wave Propagation')
    ax.legend()
    ax.grid(True)

# Create the animation
ani = FuncAnimation(fig, update, frames=range(nt), interval=100)

# Display the animation
plt.close() # to prevent it from being displayed twice
from IPython.display import HTML
HTML(ani.to_jshtml())
plt.show()
```
