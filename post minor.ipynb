```python
from sympy import symbols, integrate
x,y,c=symbols('x y c')
M=(3*(x**2)*(y**3))-5*(x**4)
N=(y+3*(x**3)*(y**2))
dM_dy=M.diff(y)
dN_dx=N.diff(x)
print("dm/dy:", dM_dy)
print("dn/dx:", dN_dx)
if(dM_dy==dN_dx):
   print("exact DE")
else:
    print("not an exact DE")
    
```

```python
I1=integrate(M,x)
I2=integrate(N.subs(x,0),y)
Gen_sol=I1+I2+c
print("general solution is:", Gen_sol)
```

```python
from sympy import symbols, integrate
x,y,c=symbols('x y c')
M=((x**2)+(3*(y**2)))
N=(2*x*y)
if(M.diff(y)==N.diff(x)):
    print("exact DE")
else:
    print("not an exact DE")
    
```

```python
from sympy import simplify,exp
exp1=simplify((M.diff(y)-N.diff(x))/N)
exp2=simplify((-M.diff(y)+N.diff(x))/M)
if(exp1.has(x) and not exp1.has(y)): #to make sure ki the exp1 has only x variableas argument and not y
    IF=exp(integrate(exp1,x))
    print("IF:", IF)
elif(exp2.has(y) and not exp2.has(x)):
    IF=exp(integrate(exp2,y))
    print("IF:", IF)
else:
    print("find IF with hit and trial")
    
```

```python
M1=M*IF
N1=N*IF
if M1.diff(y)==N1.diff(x):
    print("exactDE")
else:
    print("not an exact DE")
    
```

```python
I1=integrate(M1,x)
I2=integrate(N1.subs(x,0),y)
Gen_sol=I1+I2+c
print("general solution is:", Gen_sol)
```

```python
import matplotlib.pyplot as plt
import numpy as np
#ODE dy/dx=exp(-t)
f=lambda t,s: np.exp(-t) #define ODE, write variables, and then the equation
h=0.1 #if we reduce step size, it will give better ans, less deviation
s0=-1
t=np.arange(0,1+h,h) #arrange doesn't include the stop bound, thus using 1+h. we can use np.linspace(), using linspace we can specify the number of values we want within the range provided with both bounds included
print (t)
s=np.zeros(len(t)) #returns a new array of given space and size
s[0]=s0
for i in range(0, len(t)-1):
    s[i+1]=s[i]+h*f(t[i],s[i])
plt.figure(figsize=(8,6))
plt.plot(t,s,'bo--',label='Approximate')
plt.plot(t, -np.exp(-t), 'g',label='Exact')
plt.title('Approximate and Eaxct solutions for simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
length=10
k=0.5
t_left=100
t_right=200
tot_time=10
dt=0.01
dx=0.1
x_array=np.linspace(0,length,int(length/dx))
t_array=np.linspace(0,tot_time,int(tot_time/dt))
u=np.zeros([len(t_array),len(x_array)])
u[:,0]=t_left
u[:,-1]=t_right
for t in range(1,len(t_array)-1):
    for x in range(1,len(x_array)-1):
        u[t+1,x]=k*(dt/dx**2)*(u[t,x+1]-2*u[t,x]+u[t,x-1])+u[t,x]
plt.plot(x_array,u[-1],'black')
plt.plot(x_array,u[0])
plt.ylabel("Temperature(C)")
plt.show()
                     
```
